```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
```

```python
df = pd.read_csv('F:/upworker/Pandas Part 1/SampleSuperstore.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#descriptive analysis of the dataset
print("Statistical details of the dataset:")
print(df.describe())
print("______________________________________________________________________________________________________")

print("Summary of the dataframe:")
print(df.info())
print("______________________________________________________________________________________________________")

print("Shape of the dataset: ", df.shape)
print("______________________________________________________________________________________________________")

print("Column names in the dataset:")
print(df.columns)
print("______________________________________________________________________________________________________")

print("datatypes in the dateset: ")
print(df.dtypes)
```

```python
# Duplicate and unique values in my data

print("Duplicate values in the dataset: ", df.duplicated().sum())
print("____________________________________")
print("Unique values in the dataset:")
print(df.nunique())
```

```python
#dropping duplicate values from the dataset
df.drop_duplicates()
```

```python
#Drop the postal code column as it is not required
df = df.drop(columns = 'Postal Code', axis = 1)
```

```python
# correlation matrix of the data
df.corr()
```

```python
# covariance matrix
df.cov()
```

```python
sns.pairplot(df, hue = 'Segment')
```

```python
cl = ['Sales', 'Quantity', 'Discount', 'Profit']
for columns in cl:
    df.boxplot(column = columns, rot = 20, fontsize = 15)
    plt.show()
```

```python
for columns in cl:
    sns.distplot(df[columns], kde = True, color = 'green')
    plt.show()
```

```python
#pivot view for more exploration

df.pivot_table(index = ['Sub-Category'],
              values = ['Sales'],
              columns = ['Region', 'Segment', 'Category'],
              aggfunc = [np.sum],
              fill_value = 0, margins = True)
```

```python
df_pivot =df.pivot_table(index = ['State'],
              values = ['Sales'],
              columns = ['Region', 'Segment', 'Category'],
              aggfunc = [np.sum],
              fill_value = 0, margins = True)

df_pivot
```

```python
# total sales grouped by cities
total_sales = df.groupby('City')['Sales'].sum()

#top 5 cities by sales
top_5_cities = total_sales.sort_values(ascending = False).iloc[0:5]

#plot top 5 cities by Sales
top_5_cities.plot(kind = 'barh')
plt.title('Top 5 cities by sales')
plt.xlabel('Sales')
plt.show()

#total profit grouped by cities
total_profit = df.groupby('City')['Profit'].sum()

#top 5 cities by profit
top_5_cities = total_profit.sort_values(ascending= False).iloc[0:5]

#plot top 5 cities by Profit
top_5_cities.plot(kind = 'barh')
plt.title('Top 5 cities by Profit')
plt.ylabel('Profit')
```

```python
# total sales grouped by Region
total_sales = df.groupby('Region')['Sales'].sum()

region_sales = total_sales.sort_values(ascending = False).iloc[0:4]

#plot top 5 cities by Sales
region_sales.plot(kind = 'barh')
plt.title('Sales by Region')
plt.xlabel('Sales')
plt.show()

#total sales grouped by Segment
total_profit = df.groupby('Segment')['Sales'].sum()

segment_sales = total_profit.sort_values(ascending= False).iloc[0:5]

#Sales by different segments
segment_sales.plot(kind = 'barh')
plt.title('Sales by Segment')
plt.ylabel('Sales')
```

```python
columns = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']

for col in columns:
    sns.countplot(x= col, palette = 'icefire', data = df)
    plt.title(col, fontsize = 20)
    fig = plt.gcf()
    fig.set_size_inches(20,5)
    plt.show() 
```

```python
rcParams['figure.figsize'] = 15,5
sns.set(font_scale = 1.4)

#Average profit by region grouped by segment.
ax = sns.barplot(x = 'Region', y = 'Profit', hue = 'Segment', data = df, ci = None, palette='cubehelix_r')
ax.set_title("Average profit by segment for each region")
plt.axhline(y = df['Profit'].mean(), color = 'k', linestyle = '--')
plt.axhline(y = 0, color = 'k', linestyle = '--' )
plt.legend(loc = 'upper left')
plt.xlabel("")
plt.ylabel("Average profit")
plt.show()

#Average profit by region grouped by category.
ax = sns.barplot(x = 'Region', y = 'Profit', hue = 'Category', data = df, ci = None, palette = 'gist_stern_r')
ax.set_title('Average profit by category for each region')
plt.axhline(y = df['Profit'].mean(), color = 'k', linestyle = '--')
plt.axhline(y = 0, color = 'k', linestyle = '--')
plt.legend(loc = 'upper left')
plt.xlabel("")
plt.ylabel("Average profit")
plt.show()
```

```python
# what are the relations between discount and other features?

#Average profit by discount
ax = sns.barplot(x = 'Discount', y = 'Profit', data = df, palette = 'cubehelix_r', ci = None )
ax.set_title('Average profit by Discount', fontsize = 18)
plt.axhline(y = df['Profit'].mean(), color='k', linestyle='--')
plt.axhline(y = 0, color='k', linestyle='-')
plt.xlabel("")
plt.ylabel("Average profit")
plt.show()

#Average sales by discount
ax = sns.barplot(x = 'Discount', y = 'Sales', data = df, palette = 'gist_stern_r', ci = None )
ax.set_title('Average Sales by Discount', fontsize = 18)
plt.axhline(y = df['Sales'].mean(), color='k', linestyle='--')
plt.axhline(y = 0, color='k', linestyle='-')
plt.xlabel("")
plt.ylabel("Average Sales")
plt.show()
```

We can clearly see that sales made with less than 20% discounts are more profitable. 
more discount implies more sales but negative profits.
